```python
import numpy as np
import time
```

```python
size=1_000_000 #large data set of 1 milion

py_list=list(range(size))
start=time.time()

list_sq = [x**2 for x in py_list]

end = time.time()
print("python list time :", end-start,"second")

# numpy array
np_arr=np.array(py_list)
start=time.time()

arr_sq= np_arr ** 2 #vecroeized operation
end=time.time()
print("numpy array time :", end-start,"second")

```
