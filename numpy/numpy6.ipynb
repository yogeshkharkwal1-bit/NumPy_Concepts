```python
# data type downcastinf\g
"""we use the downcast method for save our storage in accorindg to our data
int8 to int128 data type contains memory in incresed way ,
1..means int8 take low memory storage
2..int 128 take large mem storage"""

# example- 
"""
* when we have a short int,float value we define the data type into int8
* and when we have a large value 25726738  like this we store it to int64
"""


arr=np.array([2,4,5,6,7])
print(arr,arr.dtype)

arr=np.array([2,4,5,6,7],dtype="int16")
print(arr,arr.dtype)
```
