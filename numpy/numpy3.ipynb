```python
import numpy as np
```

```python
# create in build in function

# for zero metrix
arr1=np.zeros((2,3),dtype="int64")
print(arr1,arr1.shape)

# for one  metrix
arr2=np.ones((2,3),dtype="int64")
print(arr2,arr2.shape)    

# for fixed value  metrix like 100 in every row and cllum
arr3=np.full((3,4),100,dtype="int64")
print(arr3,arr3.shape)

# for identity  metrix only digonal values are 1
arr4=np.eye((3),dtype="int64")
print(arr4,arr4.shape)

# for range values print metrix
arr5=np.arange(1,11)
print(arr5,arr5.shape)

# for range in even metrix
arr6=np.arange(1,11,2)
print(arr6,arr6.shape)

# for range but we define how values in the metrix (4 values)
arr7=np.linspace(1,100,4)
print(arr7,arr7.shape)
```
