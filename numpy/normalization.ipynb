Normalization- normalizing an array means transforming its valus so they fill into a specific scale of range

```python
import numpy as np
```

std normalization formula..

(arr-mean)/std dev

{[(VAL_i - mean)^2]/N }^1/2

```python
arr=np.array([[1,2],[3,4]])

mean=np.mean(arr)
std_dev= np.std(arr)

print(mean)
print(std_dev)

nm_arr= (arr-mean)/std_dev
print(nm_arr)
```

```python
# mathematical function

arr=np.array([1,2,3,4,5])

# agrication function

print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.max(arr))
print(np.mean(arr))
print(np.std(arr))
print(np.var(arr))
print(np.argmin(arr))
print(np.median(arr))
print(np.argmax(arr))

```
